```python
continents_list = ['아시아주','미주','구주', '대양주','아프리카주','기타대륙', '교포소계']
continents_list
```

- 국적 데이터에서 continents_list의 대륙 이름이 없는 것만 필터링

```python
kto_201901_country = kto_201901[(kto_201901['국적'].isin(continents_list)) == False]
kto_201901_country
```

```python
kto_201901_country['국적'].unique()
```

```python
kto_201901_country_newindex = kto_201901_country.reset_index(drop=True)  # 기존 인덱스 삭제하고 0부터 다시 재배열
kto_201901_country_newindex
```

- 아시아, 아메리카, 유럽, 오세아니아, 아프리카, 기타대륙, 교포 라는 이름으로 국적 옆에 대륙 삽입


```python
continents = ['아시아'] *25 + ['아메리카'] * 5 + ['유럽'] * 23 + \
['오세아니아']*3+ ['아프리카'] * 2 +['기타대륙'] +['교포']

print(len(continents))
continents
```

```python
kto_201901_country_newindex['대륙'] = continents
kto_201901_country_newindex.head()
```

- 각 국가별 국내에 관광을 위해 입국하는 관광객비율(%) 컬럼 생성
> 관광객비율(%) = 관광/계 * 100

```python
kto_201901_country_newindex['관광객비율(%)'] = round(kto_201901_country_newindex['관광']/\
                                                     kto_201901_country_newindex['계'] * 100,1)
kto_201901_country_newindex.head()
```

```python
kto_201901_country_newindex.sort_values(by='관광객비율(%)', ascending=False).head()
```

- 대륙별 관광객비율(%)의 평균

```python
kto_201901_country_newindex.pivot_table(values = '관광객비율(%)',
                                       index = '대륙',
                                       aggfunc = 'mean')
```

```python
kto_201901_country_newindex[kto_201901_country_newindex['국적']=='중국']
```

- 전체국가의 관광객 수 대비 국가 별 관광 비율
> 컬럼명 : 전체비율(%)

```python
kto_201901_country_newindex['전체비율(%)'] = round(kto_201901_country_newindex['관광']/\
                                                   kto_201901_country_newindex['관광'].sum() * 100,1)
kto_201901_country_newindex.head()
```
