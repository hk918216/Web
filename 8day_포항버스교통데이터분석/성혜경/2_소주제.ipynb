```python
df_total = pd.read_csv('C:/Users/admin/STUDY/Web/8day_포항버스교통데이터분석/포항/df_total.csv')
```

```python
df_total[df_total['환승여부'] == 'Y']
```

```python
trans = df_total[['승차정류장','환승여부']]
trans
```

- 정류장별 환승빈도 구하기

```python
trans30 = trans[trans['환승여부']=='Y'].groupby('승차정류장').count().reset_index().sort_values('환승여부',ascending=False).reset_index(drop=True).head(30)
trans30.columns = ['승차정류장','환승빈도']
```

```python
trans30
```

```python
import matplotlib.pyplot as plt
```

```python
k = []
for x in range(len(trans30)):
    k.append(trans30['승차정류장'][x])
k
```

```python
# 그래프 크기 지정하기 = 너비, 높이
plt.rcParams['figure.figsize'] = (20, 10)
# 선의 두께 지정
plt.rcParams['lines.linewidth'] = 2
# 폰트지정
plt.rc("font", family = "Malgun Gothic", size=13)

# 데이터 넣기
trans30.plot(kind='bar')

plt.title("환승을 가장 많이하는 정류장 top30")
plt.ylabel("환승빈도")

# x축 범위?
a = list(range(0,29,1))
plt.xticks([i for i in range(0,30,1)], k, rotation = 55)

# 범례 위치 변경하기
plt.legend(loc="upper right")
# 그래프 내에 그리드선 표시하기
plt.grid(True)

plt.show()
```

## 버스 내 체류시간이 가장 긴 노선명

```python
df_total.groupby('노선명').mean()
```

```python
df_total['승차시각'] = df_total['승차시각'].astype(str)
df_total['하차시각'] = df_total['하차시각'].astype(str)
```

```python
df_total['소요시간'] = pd.to_datetime(df_total['하차시각']) - pd.to_datetime(df_total['승차시각'])
df_total.head()
```

```python
df_total['소요시간(초)'] = df_total['소요시간'].astype('timedelta64[s]')
```

```python
df_total_num = df_total.groupby('노선명').sum().reset_index()
df_total_num
```

- 노선별 소요시간 합계

```python
df_total_num_dec_30 = df_total_num.sort_values('소요시간(초)', ascending = False).head(30).reset_index(drop=True)[['노선명','소요시간(초)']]
df_total_num_dec_30
```

```python
k = []
for x in range(len(df_total_num_dec_30)):
    k.append(df_total_num_dec_30['노선명'][x])
k
```

```python
# 그래프 크기 지정하기 = 너비, 높이
plt.rcParams['figure.figsize'] = (20, 10)
# 선의 두께 지정
plt.rcParams['lines.linewidth'] = 2
# 폰트지정
plt.rc("font", family = "Malgun Gothic", size=13)

# 데이터 넣기
df_total_num_dec_30.plot(kind='bar')

plt.title("가장많이 체류하는 노선 top30")
plt.ylabel("체류시간")

# x축 범위?
a = list(range(0,29,1))
plt.xticks([i for i in range(0,30,1)], k, rotation = 55)

# 범례 위치 변경하기
plt.legend(loc="upper right")
# 그래프 내에 그리드선 표시하기
plt.grid(True)

plt.show()
```
