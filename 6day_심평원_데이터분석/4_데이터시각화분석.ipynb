## 데이터 시각화

```python
import pandas as pd
```

```python
df_month_age_any = pd.read_excel('./files_new2/df_month_age_any.xlsx')
df_month_age_any.describe()
```

```python
# 월별 연령대별 수진자연령 집계하기
pivot_month_age= df_month_age_any.pivot_table(values = '수진자연령',
                            index = '요양개시월',
                            columns = '연령대',
                            aggfunc = 'count')
pivot_month_age = pivot_month_age.fillna(0)
pivot_month_age
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.rdParams['figure.figsize'] = (20,10)

plt.rdParams['lines.linewidth'] = 2

plt.rc('font', family = 'Malgun Gothic', size = 10)

pivot_month_age.piot()

plt.title('연령 및 월별로 병원을 많이 찾는 연령대 분석')
plt.ylabel('연령별 수진자수')
```
